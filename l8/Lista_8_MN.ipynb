#### Zadanie 1:

Oblicz wartości i wektory własne macierzy

$$\begin{pmatrix} -1& -4& 1\\
-1& -2& -5\\
5 & 4& 3\end{pmatrix}$$


#### Rozwiązanie:

```python
import numpy as np

# Macierz
A = np.array([[-1, -4, 1],
              [-1, -2, -5],
              [5, 4, 3]])

# Obliczenia wartości i wektorów własnych
eigenvalues, eigenvectors = np.linalg.eig(A)

# Wyświetlenie wyników
print("Wartości własne:")
print(eigenvalues)

print("\nWektory własne:")
print(eigenvectors)
```

#### Rozwiązanie:

#### Zadanie 3:

Oblicz trzy najmniejsze wartości własne i odpowiadające im wektory własne macierzy trójdiagonalnej $n\times n$

$$\begin{pmatrix} 2& -1& 0 & \dots & 0 \\
-1& 2& -1 & \dots & 0\\
0 & -1 & 2& \dots & 0\\
\vdots & \vdots & \ddots & \ddots & \vdots\\
0 & 0 & \dots -1 & 2
\end{pmatrix}$$


dla $n=10$ i $100$

#### Rozwiązanie:

```python
import numpy as np
from scipy.sparse import diags
from scipy.sparse.linalg import eigsh

def tridiagonal_matrix(n):
    # Tworzenie macierzy trójdiagonalnej
    diagonals = [2 * np.ones(n), -1 * np.ones(n - 1), -1 * np.ones(n - 1)]
    A = diags(diagonals, [0, 1, -1], format='csr')
    return A

# Rozmiary macierzy
n_values = [10, 100]

for n in n_values:
    # Tworzenie macierzy trójdiagonalnej
    A = tridiagonal_matrix(n)

    # Obliczenia trzech najmniejszych wartości własnych i odpowiadających wektorów własnych
    eigenvalues, eigenvectors = eigsh(A, k=3, which='SM')

    # Wyświetlanie wyników
    print(f"\nDla n={n}:")
    print("Najmniejsze wartości własne:")
    print(eigenvalues)

    print("\nOdpowiadające wektory własne:")
    print(eigenvectors)
```

#### Zadanie 4:

Znajdź cztery najmniejsze wartości własne kwantowego oscylatora anharmonicznego, którego równanie Schrodingera ma postać

$$-\frac{1}{2}\frac{d^2\psi}{dx^2}+\left(\frac{x^2}{2}+\lambda x^4\right)\psi(x)=E\psi(x)$$

z następującymi warunkami brzegowymi $y(0) = 0$ i $y(\pi)=0$. 


**Wskazówka:** Wprowadzając siatkę równoodległych punktów
$$x_i=-a+ih,\quad i=0,1,\dots\,,m;\quad h=\frac{2a}{m}$$
i zastępując drugą pochodną $\psi$ w $x_i$ przybliżeniem różnicy centralnej

$$\frac{d^2\psi}{dx^2}\approx\frac{\psi(x_{i-1})-2\psi(x_i)+\psi(x_{i+1})}{h^2}=\frac{\psi_{i-1}-2\psi_i+\psi_{i+1}}{h^2}$$

oraz zakładając, że $\psi_0=\psi_m=0$ możemy równanie Schrodingera zapisać w postaci macierzowej

$$\boldsymbol{H\psi}=E\boldsymbol{\psi}\,,$$

gdzie

$$\boldsymbol{\psi}=(\psi_1,\psi_2,\dots,\psi_{m-1})^T$$

i

$$\boldsymbol{H}=\begin{pmatrix} d_1& g& 0 & \dots & 0 \\
g& d_2& g & \dots & 0\\
0 & g & d_3& \dots & 0\\
\vdots & \vdots & \ddots & \ddots & \vdots\\
0 & 0 & \dots & g & d_{m-1}
\end{pmatrix}$$

z

$$g=-\frac{1}{2h^2}\,,\quad d_i=\frac{1}{h^2}+\frac{x_{i}^2}{2}+\lambda x_i^4$$

Oblicz wartości własne trójdiagonalnej macierzy $\boldsymbol{H}$ dla $\lambda=0.2$, $a=4.6$, $m=100$ i $1000$. Wykreśl 4 funkcje falowe (wektory własne $\boldsymbol{H}$) dla czterech najmniejszych energii własnych.

#### Rozwiązanie:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import eig_banded

def anharmonic_oscillator(lmbda, a, m):
    h = 2 * a / m
    x = np.linspace(-a, a, m+1)[1:-1]
    g = -1 / (2 * h**2)
    d = 1 / h**2 + (x**2) / 2 + lmbda * x**4

    H = np.zeros((3, m-1))
    H[0, 1:] = g
    H[1, :] = d
    H[2, :-1] = g

    eigenvalues, eigenvectors = eig_banded(H, lower=True)

    # Sortowanie wartości i wektorów własnych
    sorted_indices = np.argsort(eigenvalues)
    eigenvalues = eigenvalues[sorted_indices]
    eigenvectors = eigenvectors[:, sorted_indices]

    return x, eigenvalues[:4], eigenvectors[:, :4]

# Parametry
lmbda = 0.2
a = 4.6
m_values = [100, 1000]

# Wykresy funkcji falowych
plt.figure(figsize=(12, 8))

for m in m_values:
    x, energies, wavefunctions = anharmonic_oscillator(lmbda, a, m)

    for i in range(4):
        plt.subplot(2, 2, i+1)
        plt.plot(x, wavefunctions[:, i], label=f'E={energies[i]:.4f}')

# Ustawienia wykresów
plt.subplot(2, 2, 1)
plt.title('Funkcje falowe dla m=100')
plt.xlabel('x')
plt.ylabel('Psi(x)')
plt.legend()

plt.subplot(2, 2, 2)
plt.title('Funkcje falowe dla m=1000')
plt.xlabel('x')
plt.ylabel('Psi(x)')
plt.legend()

plt.subplot(2, 2, 3)
plt.xlabel('x')
plt.ylabel('Psi(x)')
plt.legend()

plt.subplot(2, 2, 4)
plt.xlabel('x')
plt.ylabel('Psi(x)')
plt.legend()

plt.tight_layout()
plt.show()
```
