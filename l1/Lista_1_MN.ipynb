$$f(x)=\cos x-3\sin(\textrm{tg}\,x -1)$$

#### Zadanie 3:

Wykorzystaj moduły *numpy* i *scipy* do działań na niżej podanych Mach
i wektorze.

$$A=\begin{pmatrix}4 & -2 & 1\\-2 & 4 & -2\\1 & -2 & 4\end{pmatrix}\,,\quad B=\begin{pmatrix}4 & 2 & 0\\2 & 5 & 2\\1 & -2 & 4\end{pmatrix}\,,\quad w=\begin{pmatrix}1 \\-2\\4\end{pmatrix} $$

Oblicz $AB$, $Aw$, $B(Aw)$ oraz wyznaczniki i Me odwrotne $A$ i $B$.

#### Rozwiązanie:

```python
import numpy as np

A = ([4, -2, 1], [-2, 4, -2], [1, -2, 4])
B = ([4, 2, 0], [2, 5, 2], [1, -2, 4])
w = ([1], [-2], [4])

print("A B")

print(np.matmul(A, B), '\n')

print("Aw")

print(np.matmul(A, w), '\n')

print("B(Aw)")

print(np.matmul(B, (np.matmul(A, w))), '\n')

print("wyznacznik(A) =", np.linalg.det(A))

print("wyznacznik(B) =", np.linalg.det(B), '\n')

print("M odwrotna A:")

print(np.linalg.inv(A), '\n')

print("M odwrotna B:")

print(np.linalg.inv(B))
```

#### Zadanie 4:

Elementy My Hilberta $n\times n$ określone są następująco:

$$h_{ik}=\frac{1}{i+k+1}\,,\quad i,k=0,\ 1,...,\ n-1\,.$$

Wypisz elementy tej My oraz My odwrotnej dla $n$=4 i 8.  

Sprawdź jak zmienia się wyznacznik tej My dla $n$ od 5 do 20.

#### Rozwiązanie:

```python
import numpy as np

def funkcja(n):
    x = []
    M = []

    for i in range(0, n):
        for k in range(0, n):
            x.append(1 / (i + k + 1))

        M.append(x)
        x = []
    return(M)
#macierz hilberta
def prt(x):
    print("Macierz Hilberta dla n =", len(x[0]), '\n')
    print(np.matrix(x), '\n')


m = funkcja(4)
prt(m)
print("Macierz odwrotna \n")
print(np.linalg.inv(m), '\n')

n = funkcja(8)
prt(n)
print("Macierz odwrotna \n")
print(np.linalg.inv(n), '\n')

for a in range(5, 21):
    z = funkcja(a)
    
    print("Wyznacznik dla", a, "=", np.linalg.det(z))
```
