#### Zadanie 1:
Korzystając z modułu *matplotlib.pyplot* wykreśl funkcję  

$$f(x)=\cos x-3\sin(\textrm{tg}\,x -1)$$

i odczytaj z wykresu liczbę jej miejsc zerowych w przedziale (0, 1.5).

#### Rozwiązanie:

```python
import numpy as nu
import matplotlib.pyplot as ma

x = nu.arange(0, 1.5, 0.01)
y = nu.cos(x) - (3 * nu.sin(nu.tan(x) - 1))

ma.plot(x, y)
ma.show()
```

#### Zadanie 2:
Napisz program w *Pythonie* do obliczenia 100 kolejnych elementów ciągu

$$x_{n+1}=3.5x_n(1-x_n),\qquad n=0,\ 1,\ 2,...$$

przyjmując początkową wartość $x_0=0.1$.  

Zapisz wyniki obliczeń w postaci listy i przedstaw je na wykresie typu scatter plot.

#### Rozwiązanie:

```python
import numpy as nu
import matplotlib.pyplot as ma

ciag = nu.arange(1, 100, 1)
x = 0.1
list = []

for _ in ciag:
    x = 3.5 * x * (1 - x)
    list.append(x)

ma.scatter(ciag, list)
ma.show()
```

#### Zadanie 3:

Wykorzystaj moduły *numpy* i *scipy* do działań na niżej podanych macierzach
i wektorze.

$$A=\begin{pmatrix}4 & -2 & 1\\-2 & 4 & -2\\1 & -2 & 4\end{pmatrix}\,,\quad B=\begin{pmatrix}4 & 2 & 0\\2 & 5 & 2\\1 & -2 & 4\end{pmatrix}\,,\quad w=\begin{pmatrix}1 \\-2\\4\end{pmatrix} $$

Oblicz $AB$, $Aw$, $B(Aw)$ oraz wyznaczniki i macierze odwrotne $A$ i $B$.

#### Rozwiązanie:

```python
import numpy as nu

A = ([4, -2, 1], [-2, 4, -2], [1, -2, 4])
B = ([4, 2, 0], [2, 5, 2], [1, -2, 4])
w = ([1], [-2], [4])

print("AB")
print(nu.matmul(A, B), '\n')
print("Aw")
print(nu.matmul(A, w), '\n')
print("B(Aw)")
print(nu.matmul(B, (nu.matmul(A, w))), '\n')
print("Det(A) =", nu.linalg.det(A))
print("Det(B) =", nu.linalg.det(B), '\n')
print("Macierz odwrotna A:")
print(nu.linalg.inv(A), '\n')
print("Macierz odwrotna B:")
print(nu.linalg.inv(B))
```

#### Zadanie 4:

Elementy macierzy Hilberta $n\times n$ określone są następująco:

$$h_{ik}=\frac{1}{i+k+1}\,,\quad i,k=0,\ 1,...,\ n-1\,.$$

Wypisz elementy tej macierzy oraz macierzy odwrotnej dla $n$=4 i 8.  

Sprawdź jak zmienia się wyznacznik tej macierzy dla $n$ od 5 do 20.

#### Rozwiązanie:

```python
import numpy as nu

def fun(n):
    x = []
    macierz = []

    for i in range(0, n):
        for k in range(0, n):
            x.append(1 / (i + k + 1))

        macierz.append(x)
        x = []
    return(macierz)

def prt(x):
    print("Macierz Hilberta dla n =", len(x[0]), '\n')
    print(nu.matrix(x), '\n')


m = fun(4)
prt(m)
print("Macierz odwrotna \n")
print(nu.linalg.inv(m), '\n')

n = fun(8)
prt(n)
print("Macierz odwrotna \n")
print(nu.linalg.inv(n), '\n')

for a in range(5, 21):
    z = fun(a)
    
    print("Wyznacznik dla", a, "=", nu.linalg.det(z))
```
